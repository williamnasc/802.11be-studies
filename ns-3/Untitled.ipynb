```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Gerando dados de exemplo (substitua pelo seu próprio conjunto de dados)
x = np.random.rand(100)  # Valores aleatórios para 'x'
y = np.random.rand(100)  # Valores aleatórios para 'y'
z = np.sin(x) + np.cos(y)  # Calculando valores para 'z'

# Criando o DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'z': z})

# Criando o gráfico 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['x'], df['y'], df['z'], c='b', marker='o')

# Personalizando o gráfico (opcional)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
plt.title('Exemplo de Gráfico 3D')

# Mostrando o gráfico
plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Read data from a CSV (replace with your own data)
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

print(z_data.values)

fig = go.Figure(data=[go.Surface(z=z_data.values)])
fig.update_layout(title='Mt Bruno Elevation', autosize=False, width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Criando dados de exemplo
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Criando a figura e os subplots
fig, axs = plt.subplots(4, 3, figsize=(12, 8))

# Preenchendo cada subplot com dados
for i, ax in enumerate(axs.flat):
    ax.plot(x, y, label=f'Subplot {i+1}')
    ax.set_title(f'Subplot {i+1}')
    ax.legend()

# Ajustando o layout
plt.tight_layout()

# Mostrando o gráfico
plt.show()
```
